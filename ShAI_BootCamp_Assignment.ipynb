### Task 1
**Basic Data Exploration**: Identify the number of rows and columns in the dataset, determine the data types of each column, and check for missing values in each column.

```python
#Task 1: Basic Data Exploration
num_rows, num_columns = df.shape
print(f"Number of rows: {num_rows}, Number of columns: {num_columns}")
```

```python
data_types = df.dtypes
print(data_types)
```

```python
# Missing Values
missing_values = df.isnull().sum()
print("\nMissing Values:")
print(missing_values)
```

### Task 2
**Descriptive Statistics**: Calculate basic statistics mean, median, mode, minimum, and maximum salary, determine the range of salaries, and find the standard deviation.


```python
#Task 2: Descriptive Statistics
# Calculate basic statistics
basic_stats = df['TotalPay'].describe()
print("\nDescriptive Statistics for Salary:")
print(basic_stats)
```

```python
# range of salaries
salary_range = df['TotalPay'].max() - df['TotalPay'].min()
print(f"\nThe range of salaries : is from {df['TotalPay'].min()} to {df['TotalPay'].max()}, with a standard deviation of {salary_range}")
```

```python
#  standard deviation of salaries
salary_std = df['TotalPay'].std()
print(f"\nStandard Deviation of Salaries: {salary_std}")
```

### Task 3
**Data Cleaning**: Handle missing data by suitable method with explain why you use it.


```python
df.fillna(0, inplace=True)
```

```python
# to make sure that we fill missing values
missing_values = df.isnull().sum()
print("\nMissing Values:")
print(missing_values)
```

### Task 4
**Basic Data Visualization**: Create histograms or bar charts to visualize the distribution of salaries, and use pie charts to represent the proportion of employees in different departments.

```python
import matplotlib.pyplot as plt

# Visualize the distribution of salaries with matplotlib
plt.figure(figsize=(10, 6))
plt.hist(df['TotalPay'], bins=30, edgecolor='black', alpha=0.7)
plt.title('Distribution of Salaries')
plt.xlabel('Total Pay')
plt.ylabel('Frequency')
plt.show()

```

```python
# Represent proportion of employees in different departments with matplotlib pie chart
department_proportion = df['JobTitle'].value_counts().head(10)
plt.figure(figsize=(10, 10))
plt.pie(department_proportion, labels=department_proportion.index, autopct='%1.1f%%', startangle=90)
plt.title('Proportion of Employees in Different Departments')
plt.show()
```

### Task 5
**Grouped Analysis**: Group the data by one or more columns and calculate summary statistics for each group, and compare the average salaries across different groups.

```python

# Group data by 'Year' and calculate average salaries
average_salaries_by_year = df.groupby('Year')['TotalPay'].mean()
print(average_salaries_by_year)
```

### Task 6
**Simple Correlation Analysis**: Identify any correlation between salary and another numerical column, and plot a scatter plot to visualize the relationship.


```python
# Identify correlation between 'TotalPay' and 'Benefits'
correlation = df['TotalPay'].corr(df['Benefits'])

plt.figure(figsize=(8, 6))
plt.scatter(x=df['TotalPay'], y=df['Benefits'], alpha=0.5)
plt.title(f'TotalPay vs Benefits\nCorrelation: {correlation:.2f}')
plt.xlabel('Total Pay')
plt.ylabel('Benefits')
plt.show()
```
